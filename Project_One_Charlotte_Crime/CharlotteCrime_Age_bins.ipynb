```python
# Import Dependencies
import pandas as pd
```

```python
#Import data 
csv_path = "Resources/ois.csv" #Insert file here after downloading file
Officer_shootings_df = pd.read_csv(csv_path)
Officer_shootings_df.head()
```

```python
#creater bins
age_bins = [0, 10, 15, 20, 25, 30, 35, 40, 45, 50, 55, 60, 65, 70, 75, 80, 100]

#create group names
group_names = ["Under 10", "11-15","16-20", "21-25", "26-30", "31-35", "36-40", "41-45", "46-50", "51-55", "56-60", "61-65", "66-70", "71-75", "76-80", "Over 80"]
```

```python
#drop in bins
Officer_shootings_df["age range"]= pd.cut(Officer_shootings_df["age"], age_bins, labels=group_names)
Officer_shootings_df.head()
```

```python
#group by age bins

Officer_shootings_df_bins = Officer_shootings_df.groupby("age range")
Officer_shootings_df_bins.head()
Officer_shootings_df_bins.max()
```
